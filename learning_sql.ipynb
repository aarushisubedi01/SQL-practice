"Bookings" Table Creation

```python
%%sql
CREATE TABLE bookings (
    booking_id INTEGER PRIMARY KEY,
    customer_id INT,
    room_id INT,
    check_in DATE,
    check_out DATE,
    FOREIGN KEY(customer_id) REFERENCES customers(customer_id),
    FOREIGN KEY(room_id) REFERENCES rooms(room_id)
);
```

Displaying hotels which are in "Pokhara"

Displaying Single Rooms

```python
%%sql
SELECT room_number, hotel_id, price_per_night 
FROM rooms 
WHERE room_type = 'Single';
```

Displaying all bookings for customer "Aarav Shah"

```python
%%sql
SELECT b.booking_id, r.room_number, h.name AS hotel_name, b.check_in, b.check_out
FROM bookings b
JOIN customers c ON b.customer_id = c.customer_id
JOIN rooms r ON b.room_id = r.room_id
JOIN hotels h ON r.hotel_id = h.hotel_id
WHERE c.name = 'Aarav Shah';
```

Displaying all Payments above 15,000

```python
%%sql
SELECT * FROM payments
WHERE amount > 15000;
```

Total Customers count

```python
%%sql
SELECT COUNT(*) AS total_customers FROM customers;
```

Count of number of rooms per hotel

```python
%%sql
SELECT hotel_id, COUNT(*) AS total_rooms
FROM rooms
GROUP BY hotel_id;

```

Customers Total Payments

```python
%%sql
SELECT c.name, SUM(p.amount) AS total_paid
FROM payments p
JOIN bookings b ON p.booking_id = b.booking_id
JOIN customers c ON b.customer_id = c.customer_id
GROUP BY c.customer_id;
```

Bookings with check-in in May 2024

```python
%%sql
SELECT * FROM bookings
WHERE check_in >= '2024-05-10' AND check_out <= '2024-05-31';
```
